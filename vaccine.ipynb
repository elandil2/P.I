```python
import pandas as pd 
import numpy as np
```

```python
#imported dataset to googledrive to work on colab and previewed with describe
df = pd.read_csv('/content/drive/MyDrive/country_vaccination_stats.csv')
df.describe()
```

```python
#Question 4 -Code Implementation Task: Implement code to fill the missing data (impute) in daily_vaccinations column per country with the minimum daily vaccination number of relevant countries.  
#Note: If a country does not have any valid vaccination number yet, fill it with “0” (zero). 
#lambda for minimum values for grouped countries and if its not filled with minimum value then assign 0
df["daily_vaccinations"]=df.groupby("country")['daily_vaccinations'].transform(lambda x: x.fillna(x.min())).fillna(0)
df.head(20)
```

```python
#Question 5 - Code Implementation Task: Implement code to list the top-3 countries with highest median daily vaccination numbers by considering missing values imputed version of dataset.
#Grouped by country again  and get mean values for each country group then sort values by mean as descending and show first 3 rows
df_1 = df.groupby('country')['daily_vaccinations'].agg(['mean'])
df_1.sort_values(by='mean',ascending=False).head(3)
```
